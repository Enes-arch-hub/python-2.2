```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age":[22,np.nan,23,30,np.nan],
    "Salary":[3000,4000,None,5000,4500]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df.isnull().sum())
```

```python
df_clean = df.dropna()
print(df_clean)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
print(df)
```
